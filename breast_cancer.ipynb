```python
col_to_scale = ['radius_mean', 'texture_mean', 'perimeter_mean','area_mean','radius_se','texture_se','perimeter_se','area_se','radius_worst','texture_worst','perimeter_worst','area_worst']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[col_to_scale] = scaler.fit_transform(df[col_to_scale])
```

```python
df.head()
```

```python
dummies = pd.get_dummies(df['diagnosis'])
```

```python
dummies.head()
```

```python
df_1 = pd.concat([df, dummies], axis = 'columns')
```

```python
df_1.head()
```

```python
df_1.drop('id', axis = 'columns', inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df_1['diagnosis'] = encoder.fit_transform(df['diagnosis'])
```

```python
df_1.head()
```

```python
X = df_1.drop('diagnosis', axis = 'columns')
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 38)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB, GaussianNB
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),  # Increase max_iter to avoid convergence warnings
    'Random Forest Classifier': RandomForestClassifier(),
    'Support Vector Machine': SVC(),
    'Lasso': Lasso(max_iter=10000),  # Increase max_iter to avoid convergence warnings
    'Multinomial Naive Bayes': MultinomialNB(),
    'Gaussian Naive Bayes': GaussianNB()
}
```

```python
params = {
    'Logistic Regression': [{'penalty': ['l2'], 'C': [0.1]}],  
    'Random Forest Classifier': [{'max_depth': [None], 'n_estimators': [100]}],  
    'Support Vector Machine': [{'C': [10], 'gamma': ['scale']}], 
    'Lasso': [{'alpha': [0.1]}], 
    'Multinomial Naive Bayes': [{'alpha': [0.1]}],  
    'Gaussian Naive Bayes': [{}]  
}

grid_search_params = {'cv': 5, 'verbose': 1}
```

```python
for name, model in models.items():
    param_grid = params[name]
    grid_search = GridSearchCV(model, param_grid, **grid_search_params)
    
    
    grid_search.fit(X_train, y_train)
    
    
    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best score for {name}: {grid_search.best_score_}")
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(C = 0.1, gamma = 'scale')
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print('Classification Report: \n', classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True)
```
